```python
import requests
from bs4 import BeautifulSoup as bs 
import pandas as pd 
import numpy as np
```

```python
#get the data from tables on the page
url='https://en.wikipedia.org/wiki/PFA_Players%27_Player_of_the_Year'
page = requests.get(url)
data=page.content
soup = bs(data, 'html.parser')

players = []
for element in soup.select('table'):
    players.append(element.text.strip())
    
```

```python
#Keep the table with the wanted data (PFA)
data = players[1]
```

```python
cleaned = []
cleaned.append(data.split('\n'))
to_use = cleaned[0]
```

```python
for i in range(len(to_use)): 
    if to_use[i] == '': 
        to_use.pop(i)
```

```python
fullData = []
i = 0
while i < len(to_use):
    if to_use[i][:2] == '19'or to_use[i][:2] == '20':
        vector.append(to_use[i])
        vector.append(to_use[i+1])
        i += 1
    fullData.append(vector)
    i += 1
```

```python
data = fullData[1]
data = data[:96]
year = data[::2]
player = data[1::2]
```

```python
#Modify the year format
formated = []
for don in year:
    if don[:2] == '20': 
        don = don.replace('–','/')
        don = don[:5] + '20' + don[5:]
        formated.append(don)
    else: 
        formated.append(don)
```

```python
arr = list(zip(formated, player))
df = pd.DataFrame(arr,
                  columns = ['season', 'name'])
```
